<h2> Feature Engineering </h2>

After exploring the data, you can choose to modify features that you would use to predict the performance of the students on your chosen response variable. 

You can also create your own features. For example, perhaps you figured that maybe a state's expenditure per student may affect their overall academic performance so you create a expenditure_per_student feature.

Use this space to modify or create features.

```python
# @@@ 5

# Expenditure per student 

df["EXPENDITURE_PER_PUPIL"] = df["TOTAL_EXPENDITURE"] / df["ENROLL_ALL"]

# Instruction expenditure as a fraction of total expenditure

df["INSTRUCTION_SPENDING_RATIO"] = df["INSTRUCTION_EXPENDITURE"] / df["TOTAL_EXPENDITURE"]
```

Feature engineering justification: 

**Expenditure per student:
I created this feature by dividing total expenditure by total enrollment. This normalizes spending across states of different sizes, making comparisons fair. A large state with high total spending may not necessarily invest more in each student, so per-pupil expenditure is a better measure of how much resources are devoted to individual students.

Instruction spending ratio:
I calculated the share of total expenditure spent specifically on instruction. This shows how much of a state's education budget goes directly into classroom teaching versus other categories**

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
# @@@ 8
# X is commonly used to denote the input data
# y is used for the response / output data
X = df[['ENROLL_8','EXPENDITURE_PER_PUPIL','AVG_MATH_4_SCORE']].dropna()
y = df.loc[X.index]['AVG_MATH_8_SCORE']

y.fillna(y.median(), inplace=True)
```

```python
# @@@ 9 
# the test_size parameter defines what % of data is set aside for testing, 70 / 30 and 80 / 20 split are both typical
# we don't have a huge data set but we still want to have a decently sized testing set
# so we are using a 70 / 30 train / test split. 
# setting random_state explicitly ensures that I get the same results each time I run the code
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)

# X_train, X_test, y_train, y_test = train_test_split(
#      X, y, test_size=, random_state=42)
```
